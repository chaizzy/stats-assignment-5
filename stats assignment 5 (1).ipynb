```python
# ANS 1 :
# "estimation statistics" is a field of statistics that deals with estimation of unknown population parameters
# so, we take a sample of population and use statistical methods to estimate population mean,variance , std

# point estimate
# single numerical value to estimate the unknown population parameter

# interval estimate 
# range of values used to estimate the unknown population parameters
```

```python
# ANS 2:
# we assume confidence interval 95%
import math

def estimate_population_mean(sample_mean, sample_std_dev, sample_size):
    standard_error = sample_std_dev / math.sqrt(sample_size)
    
    margin_of_error = 1.96 * standard_error
    
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    return lower_bound, upper_bound
```

```python
# ANS 11:
# Student's t-distribution is a probability distribution used in statistical inference for hypothesis testing and 
# confidence interval estimation of the mean of a normally distributed population 
# when the sample size is small (less than 30) or the population standard deviation is unknown. 
# t-distribution is also defined by parameter called degree of freedom

# The t-distribution is used in situations where the population standard deviation is unknown and must be estimated from the sample data, 
# or when the sample size is small 
```

```python
# ANS 12:
# t-statistics is a numerical measure used in hypothesis testing to determine 
# the significance of the difference between the mean of a sample and the population mean
# t = (x̄ - μ) / (s / √n)
```

```python
# ANS 13:
import math

sample_size = 50
sample_mean = 500
sample_std = 50

z_score = 1.96  # from a standard normal distribution table

# Calculate the confidence interval
margin_of_error = z_score * (sample_std / math.sqrt(sample_size))
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Print the results
print("sample mean:",sample_mean)
print("Sample standard deviation:",sample_std)
print("95% Confidence interval" , lower_bound ,"-", upper_bound)
```

```python
# ANS 14:
import scipy.stats as stats

# null_hypothesis = new drug will not decrease the blood pressure by 10 mmhg
# alternate_hypothesis = new drug will decrease the blood pressure by 10 mmhg

sample_mean = 8
population_mean = 10
sample_std = 3
sample_size = 100
significance_level = 0.05

# Compute the test statistic
t = (sample_mean - population_mean) / (sample_std / (sample_size ** 0.5))

critical_value = stats.t.ppf(0.9750 , 99)

if abs(t) > critical_value:
    conclusion = "Reject the null hypothesis"
else:
    conclusion = "Fail to reject the null hypothesis"

print("Test statistic (t):", t)
print("Critical value:", critical_value)
print("Conclusion:", conclusion)
```

```python
# ANS 16 :
import numpy as np
import scipy.stats as stats


sample_mean_1 = 80
sample_std_1 = 10
sample_size_1 = 30

sample_mean_2 = 75
sample_std_2 = 8
sample_size_2 = 40

significance_level = 0.01

degrees_of_freedom = sample_size_1 + sample_size_2 - 2
pooled_std = np.sqrt(((sample_size_1 - 1) * sample_std_1 ** 2 + (sample_size_2 - 1) * sample_std_2 ** 2) / degrees_of_freedom)

t = (sample_mean_1 - sample_mean_2) / (pooled_std * np.sqrt(1 / sample_size_1 + 1 / sample_size_2))


critical_value = stats.t.ppf(1 - (significance_level / 2), degrees_of_freedom)


if abs(t) > critical_value:
    conclusion = "Reject the null hypothesis"
else:
    conclusion = "Fail to reject the null hypothesis"

print("Test statistic (t):", t)
print("Critical value:", critical_value)
print("Conclusion:", conclusion)
```
